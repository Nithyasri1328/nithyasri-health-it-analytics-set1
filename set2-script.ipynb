```python
import pandas as pd
import numpy as np

# Create patient DataFrame
df = pd.DataFrame({
    'Type': ['High', 'Low'],
    'Age': [30, 40]
})

print("Original Data:")
print(df)

# Clean data
df = df.fillna(0)

print("\nCleaned Data:")
print(df)

# Group by Type and calculate mean age
group_mean = df.groupby('Type')['Age'].mean()

print("\nMean Age by Type:")
print(group_mean)

# Convert Age column to NumPy array
arr = np.array(df['Age'])

print("\nNumPy Array:")
print(arr)

# Calculate average age
print("\nAverage Age:")
print(arr.mean())
```
